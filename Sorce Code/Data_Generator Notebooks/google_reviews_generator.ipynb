```python
import random
import datetime
!pip install uuid
import uuid

# Defining the values
st_date = 20230515
num_days = 2
restaurants = ['Namaste India', 'Malibu Kitchen', 'Carribean Dreams', 'Spanish Delight',
               'Greek Love', 'Little Italy', ]
restaurant_ids = ['56849', '58976', '52314', '64892',
                  '54781', '67432']
county = ['Bergen', 'Passaic', 'Passaic', 'Bergen', 'Bergen', 'Bergen']
# restaurant_addr = ['483 Sussex Street Buffalo, NY 14215', '8653C Summit Ave Brooklyn, NY 11209', '419 W Somerset Court South Ozone Park, NY 11420', '585 Southampton Dr Fresh Meadows, NY 11365',
#                    '46 Fremont Lane Brooklyn, NY 11203', '713 Elmwood Ave Bronx, NY 10467', '6 Newcastle Ave Bonx, NY 10451', '9822 SE Crescent Ave Buffalo, NY 14224', '8014 Illinois Street New York, NY 10031']
header = ['review_id', 'review_date', 'restaurant_id', 'restaurant_name',
          'county', 'rating', 'reviewer_name', 'comments', 'operation']
ratings_dist = [[3, 4, 5],[2, 3], [2, 3], [3, 4], [3, 4, 5], [4, 5]]
print(', '.join(header))

#Creating the data
start_date = datetime.datetime.strptime(str(st_date), "%Y%m%d")
restaurant_data =[]
restaurant_data.append(header)
for day in range(num_days):
    end_date = start_date + datetime.timedelta(days=day)
    en_date = end_date.strftime('%Y%m%d')
    # print(en_date)
    num_reviews = random.randint(7, 12)
    for i in range(0, num_reviews):
        id = (random.choice([0, 1, 1, 1, 1, 2, 2, 3, 3, 4, 5, 5, 5, 5 ]))

        #print(f" {en_date}, {restaurant_ids[id]}, {restaurants[id]}, {restaurant_addr[id]}, {random_rating}, ****, ***** ")

        restaurant_data.append([str(uuid.uuid1()), en_date, restaurant_ids[id], restaurants[id], county[id], random.choice(ratings_dist[id]) , '****', '*****', 'add'])

#restaurant_data[:20]
```
