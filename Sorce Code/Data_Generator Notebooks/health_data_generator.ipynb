```python
#@title
import random
import datetime
from dateutil import rrule

restaurants = ['Namaste India', 'Malibu Kitchen', 'Carribean Dreams', 'Spanish Delight',
               'Greek Love', 'Little Italy']
cuisine = ['Indian', 'Hawaiian', 'Carribean', 'Spanish', 'Greek', 'Italian']               
restaurant_ids = ['56849', '58976', '52314', '64892', '54781', '67432']
county = ['Bergen', 'Passaic', 'Passaic', 'Bergen', 'Bergen', 'Bergen']  
# restaurant_addr = ['483 Sussex Street Buffalo, NY 14215', '8653C Summit Ave Brooklyn, NY 11209', '419 W Somerset Court South Ozone Park, NY 11420', '585 Southampton Dr Fresh Meadows, NY 11365',
#                    '46 Fremont Lane Brooklyn, NY 11203', '713 Elmwood Ave Bronx, NY 10467', '6 Newcastle Ave Bonx, NY 10451', '9822 SE Crescent Ave Buffalo, NY 14224', '8014 Illinois Street New York, NY 10031']
header = ['grade_date', 'restaurant_id', 'restaurant_name', 'cuisine', 'county' ,'health_grade']
health_grade = [[3], [1, 2], [2, 3], [3, 4], [3, 4], [4, 5]]
health_data =[]
health_data.append(header)
#print(', '.join(header))
st_date = 20221101
start_date = datetime.datetime.strptime(str(st_date), "%Y%m%d")
end_date = start_date + datetime.timedelta(days=200)

# for day in range(0,120,30):
#     end_date = start_date + datetime.timedelta(days=day)
#     en_date = end_date.strftime('%Y%m%d')
for dt in rrule.rrule(rrule.MONTHLY, dtstart=start_date, until=end_date):
    en_date = dt.strftime('%Y%m%d')
    for id in range(0, 6):
        #print(f'{en_date}, {restaurant_ids[id]}, {restaurants[id]}, {restaurant_addr[id]}, {random.choice(health_grade)} ')
        health_data.append([en_date, restaurant_ids[id], restaurants[id], cuisine[id], county[id], random.choice(health_grade[id])])
print(health_data)
```
